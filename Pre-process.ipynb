```python
# All imports are stored here.
import cv2
import os
import numpy as np
from os import listdir
from PIL import Image
```

```python
# extract first fram from video and place it into output path with original video's file name.

def extractFrame(pathIn, pathOut):
    cap = cv2.VideoCapture(pathIn)
    f_name, f_ext = os.path.splitext(os.path.basename(pathIn))
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
        cv2.imwrite(os.path.join(pathOut, "{}.jpg".format(f_name)), frame)  # save frame as JPEG file
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
```

```python
# generate a dataset compatible with resnet.
# hence, a dataset of images using extractFrame function.

directory_path_in = 'D:\\Documents\\Masters\\Semester 1\\CSCM38-Advanced-Topics-AI-and-Cyber-Security\\BDDM\\Celeb-DF-v2\\YouTube-real\\'
directory_path_out = 'D:\\Documents\\Masters\\Semester 1\\CSCM38-Advanced-Topics-AI-and-Cyber-Security\\BDDM\\Celeb-DF-Image\\YouTube-real\\'
list_of_files = listdir(directory_path_in)

for video in list_of_files:
    extractFrame(directory_path_in+video, directory_path_out)
```

```python
# From the images extracted from the videos, Faces must be detected. This function will detect and crop the image to the faces
# min image size of 128 x 128.
def face_detect(PathIn, PathOut):
    cascPath = "haarcascade_frontalface_alt.xml"
    faceCascade = cv2.CascadeClassifier(cascPath)
    
    img = cv2.imread(PathIn)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    f_name = os.path.basename(PathIn)
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(128, 128),
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    for (x, y, w, h) in faces:
        roi_color = img[y:y + h, x:x + w]
        cv2.imwrite(os.path.join(PathOut, f_name), roi_color)
```

```python
directory_path_in = "./Celeb-DF-Image/Celeb-real/"
directory_path_out = "./Celeb-DF-Image-Face-Crop/Celeb-real/"
list_of_files = listdir(directory_path_in)
#print(list_of_files)
for image in list_of_files:
    face_detect(directory_path_in+image, directory_path_out)
```

```python
# Celeb-df only provides a partial label set calles test_labels. Using it's format, an entire label set is generated called
# Full_labels.txt (this is for the image dataset generated only).

def generate_labels(directory_path_in, filename, subfolder, value):
    list_of_files = listdir(directory_path_in)
    f = open(filename, "a")
    for image in list_of_files:
        f.write("{v} {s}{i}\n".format(v=value, s=subfolder, i=image))
    f.close()
```

```python
# The SRM filter will be applied onto to the dataset
# The SRM kernel is defined below.

SRM_kernel = np.array([[-1, 2, -2, 2, -1],
                    [2, -6, 8, -6, 2],
                    [-2, 8, -12, 8, -2],
                    [2, -6, 8, -6, 2],
                    [-1, 2, -2, 2, -1]])
SRM_kernel = SRM_kernel/12
```

```python
def apply_kernel(PathIn, PathOut):
    img = cv2.imread(PathIn)
    #filter the source image
    img_rst = cv2.filter2D(img,-1,SRM_kernel)
    #save result image
    f_name = os.path.basename(PathIn)
    cv2.imwrite(os.path.join(PathOut, f_name),img_rst)
```

```python
directory_path_in = "./Celeb-DF-Image-Face-Crop/Celeb-synthesis/"
directory_path_out = "./Celeb-DF-Image-Face-Crop-SRM/Celeb-synthesis/"
list_of_files = listdir(directory_path_in)
#print(list_of_files)
for image in list_of_files:
    apply_kernel(directory_path_in+image, directory_path_out)
```
